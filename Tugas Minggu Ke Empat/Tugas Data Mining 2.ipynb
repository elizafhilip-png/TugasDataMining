```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Membaca dataset
df = pd.read_csv("RealEstate.csv")

# =============================
# EKSPLORASI DATA
# =============================

print("===== 5 DATA PERTAMA =====")
print(df.head())

print("\n===== INFO DATA =====")
print(df.info())

print("\n===== STATISTIK DESKRIPTIF =====")
print(df.describe())

print("\n===== CEK DATA KOSONG =====")
print(df.isnull().sum())


# =============================
# VISUALISASI DATA
# =============================

# 1 Histogram distribusi harga rumah
plt.figure(figsize=(8,5))
plt.hist(df["Price"], bins=30)
plt.title("Distribusi Harga Rumah")
plt.xlabel("Price")
plt.ylabel("Jumlah Rumah")
plt.show()


# 2 Bar Chart jumlah rumah berdasarkan lokasi
plt.figure(figsize=(10,6))
df["Location"].value_counts().plot(kind="bar")
plt.title("Jumlah Rumah Berdasarkan Lokasi")
plt.xlabel("Location")
plt.ylabel("Jumlah Rumah")
plt.show()


# 3 Bar Chart jumlah rumah berdasarkan status
status_counts = df["Status"].value_counts()

plt.figure(figsize=(8,6))
plt.bar(status_counts.index, status_counts.values)

plt.xlabel("Status")
plt.ylabel("Jumlah Rumah")
plt.title("Jumlah Rumah Berdasarkan Status")

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


# 4 Pie Chart persentase status rumah
plt.figure(figsize=(7,7))
df["Status"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.title("Persentase Status Rumah")
plt.ylabel("")
plt.show()


# 5 Scatter plot hubungan luas rumah dan harga
plt.figure(figsize=(8,6))
plt.scatter(df["Size"], df["Price"])
plt.title("Hubungan Luas Rumah dan Harga")
plt.xlabel("Size")
plt.ylabel("Price")
plt.show()


# 6 Scatter hubungan bedrooms dan price
plt.figure(figsize=(8,6))
sns.scatterplot(x="Bedrooms", y="Price", data=df)
plt.title("Hubungan Bedrooms dan Price")
plt.show()


# 7 Boxplot harga rumah
plt.figure(figsize=(8,6))
sns.boxplot(x=df["Price"])
plt.title("Boxplot Harga Rumah")
plt.show()


# 8 Boxplot harga berdasarkan status
plt.figure(figsize=(8,6))
sns.boxplot(x="Status", y="Price", data=df)
plt.title("Harga Rumah Berdasarkan Status")
plt.show()


# 9 Line chart rata-rata harga berdasarkan jumlah kamar
avg_price = df.groupby("Bedrooms")["Price"].mean()

plt.figure(figsize=(8,6))
avg_price.plot(kind="line", marker="o")
plt.title("Rata-rata Harga Berdasarkan Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Average Price")
plt.show()


# 10 Countplot jumlah rumah berdasarkan status
plt.figure(figsize=(8,6))
sns.countplot(x="Status", data=df)
plt.title("Jumlah Rumah Berdasarkan Status")
plt.show()


# 11 Heatmap korelasi antar variabel
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Heatmap Korelasi Variabel")
plt.show()


# =============================
# MULTIPLE SUBPLOT (CONTOH)
# =============================

t = np.arange(0,10)

plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
plt.plot(t,t**2,color='red')
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = x^2")

plt.subplot(2,2,2)
plt.plot(t,t**3,color='blue')
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = x^3")

plt.tight_layout()
plt.show()


# Plot beberapa fungsi dalam satu grafik
plt.plot(t, t,'g',
         t, t**2,'bs',
         t, t**3,'r')
plt.title("Contoh Multiple Line Plot")
plt.show()
```
