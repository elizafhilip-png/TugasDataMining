```python
import pandas as pd
from google.colab import files

# Upload file CSV
print("/kaggle/archive.zip")
uploaded = files.upload()

# Ambil nama file otomatis
file_name = list(uploaded.keys())[0]

# Membaca file
df = pd.read_csv(file_name)

# Menampilkan isi data
print("=== DATA KELAYAKAN PENDIDIKAN INDONESIA ===")
print(df)

print("\n=== 5 DATA PERTAMA ===")
print(df.head())

print("\n=== 5 DATA TERAKHIR ===")
print(df.tail())

print("\n=== INFO DATA ===")
print(df.info())

print("\n=== STATISTIK DESKRIPTIF ===")
print(df.describe())

print("\n=== NAMA KOLOM ===")
print(df.columns)

print("\n=== JUMLAH MISSING VALUE ===")
print(df.isnull().sum())
```
