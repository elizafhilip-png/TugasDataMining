```python
# ================================
# DATA PREPROCESSING
# ================================

# 1. Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()
```

Pada tahap ini, dilakukan import library yang dibutuhkan. Library pandas digunakan untuk mengolah data dalam bentuk tabel (dataframe), sedangkan numpy digunakan untuk membantu operasi numerik.

```python
# 2. Load Dataset
df = pd.read_csv('student_performance.csv')

print("=== Data Awal ===")
print(df.head())
```

Dataset dibaca menggunakan fungsi read_csv(). Selanjutnya, fungsi head() digunakan untuk menampilkan 5 data pertama agar dapat mengetahui gambaran awal isi dataset.

```python
# 3. Informasi Data
print("\n=== Info Data ===")
print(df.info())
```

Fungsi info() digunakan untuk melihat struktur dataset, seperti jumlah data, nama kolom, tipe data, serta apakah terdapat nilai kosong (missing value).

```python
# 4. Statistik Deskriptif
print("\n=== Statistik Data ===")
print(df.describe())
```

Fungsi describe() digunakan untuk menampilkan statistik dasar dari data numerik, seperti nilai rata-rata (mean), nilai minimum, maksimum, dan standar deviasi.

```python
# 5. Cek Missing Value
print("\n=== Missing Value ===")
print(df.isnull().sum())
```

Tahap ini bertujuan untuk mengecek jumlah data yang kosong pada setiap kolom. Missing value perlu ditangani agar tidak mempengaruhi hasil analisis.

```python
# 6. Mengatasi Missing Value
# Numerik → isi dengan mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# Kategorik → isi dengan modus
for col in df.select_dtypes(include='object'):
    df[col].fillna(df[col].mode()[0], inplace=True)

print("\n=== Missing Value Setelah Ditangani ===")
print(df.isnull().sum())
```

Missing value pada data numerik diisi menggunakan nilai rata-rata (mean), sedangkan pada data kategorik diisi menggunakan nilai yang paling sering muncul (modus). Hal ini dilakukan agar data menjadi lebih lengkap tanpa menghilangkan terlalu banyak informasi.

```python
# 7. Cek Duplikasi
print("\n=== Data Duplikat ===")
print(df.duplicated().sum())
```

Digunakan untuk mengecek apakah terdapat data yang duplikat atau sama persis. Data duplikat dapat menyebabkan bias dalam analisis sehingga perlu diidentifikasi.

```python
# 8. Encoding Data Kategorik
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in df.select_dtypes(include='object'):
    df[col] = le.fit_transform(df[col])

print("\n=== Data Setelah Encoding ===")
print(df.head())
```

Data kategorik (berupa teks seperti “male”, “female”) diubah menjadi bentuk numerik menggunakan LabelEncoder. Hal ini diperlukan karena algoritma machine learning hanya dapat memproses data dalam bentuk angka.

```python
# 9. Normalisasi Data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

print("\n=== Data Setelah Normalisasi ===")
print(df_scaled.head())

# 10. Simpan Hasil
df_scaled.to_csv('data_preprocessing.csv', index=False)

print("\nPreprocessing selesai! Data disimpan sebagai 'data_preprocessing.csv'")
```

Normalisasi dilakukan menggunakan MinMaxScaler untuk mengubah rentang nilai data menjadi 0 sampai 1. Tujuannya agar setiap fitur memiliki skala yang sama sehingga tidak ada fitur yang mendominasi. Dataset yang telah melalui proses preprocessing disimpan kembali dalam bentuk file CSV agar dapat digunakan untuk tahap selanjutnya.
